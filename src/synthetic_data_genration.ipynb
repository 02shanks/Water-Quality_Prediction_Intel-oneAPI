##load_data

```python
from ctgan import CTGAN
from ctgan import load_demo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
csv = "/content/drive/MyDrive/Colab_Notebooks/intel_oneapi_water/balanced_data_water.csv"
```

```python
import pandas as pd
df =pd.read_csv(csv)
df.head()
```

```python
df.info()
```

```python
# Names of the columns that are discrete
discrete_columns = [
'Target',
'Color_Colorless' ,
'Color_Faint Yellow',
'Color_Light Yellow',
'Color_Near Colorless',
'Color_Yellow',
'Day',
'Time of Day'
]
```

```python
params = {'verbose': True,
 'epochs': 300,
 'pac': 10,
 'cuda': True}
ctgan = CTGAN(**params)
```

```python
ctgan.fit(df, discrete_columns)
```

```python
# Create synthetic data
num_sample = 1000000
synthetic_data = ctgan.sample(num_sample)
```

```python
synthetic_data.to_csv("/content/drive/MyDrive/Colab_Notebooks/intel_oneapi_water/ctgan_syn_data_water.csv", index=False)
```
